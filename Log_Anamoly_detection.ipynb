```python
import pandas as pd
# !pip3 install ijson
import ijson

data = ijson.parse(open("/kaggle/input/convolve-epoch1/train.json", 'r'))

num = 4
ctr = 0
dummy ={}
for prefix, event, value in data:
    dummy[prefix] = value
    ctr+=1
    if(ctr > num):
        break
del dummy['']
print(dummy)
```
